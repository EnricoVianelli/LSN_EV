### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

<span style="color:red">Show a picture of both your estimations of $I$</span> and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$</font>.

```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from file
data = np.genfromtxt("Integrali-1.out")

# Extract values and uncertainties
blocks = data[:, 0]  # Number of blocks
integral_values = data[:, 1]  # Integral values
uncertainties = data[:, 2]  # Uncertainties

# Plotting
plt.figure(figsize=(12,7))
plt.errorbar(blocks, integral_values, yerr=uncertainties, fmt='o', markersize=3, capsize=5, color='magenta')
plt.xlabel('Number of blocks')
plt.ylabel('$I$')
plt.axhline(y=1.0, color='b', linestyle='--', label='True I')
plt.title('Performing Monte Carlo sampling uniform distribution [0,1]')
plt.grid(True)
plt.show()
```

Below i tried importance sampling, using the following function:
$$f(x) = 2(1-x) $$ 
Note that $f(x)$ is already normolized: $\int_0^1f(x)dx = 1 $

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f1(x):
    return (np.pi / 2) * np.cos(np.pi * x / 2)

def f2(x):
    return 2 * (1 - x)

# Generate x values in the region [0, 1]
x = np.linspace(0, 1, 500)

# Calculate y values for both functions
y1 = f1(x)
y2 = f2(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r"$\frac{\pi}{2}\cos(\frac{\pi x}{2})$", linewidth=2)
plt.plot(x, y2, label=r"$2(1-x)$", linewidth=2, linestyle='--')

# Add labels, legend, and grid
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.title("Graphs of the Functions in [0, 1]", fontsize=16)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$
2. in the continuum; at each discrete time the walker makes a step of length equal to $a(=1)$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

<span style="color:red">Show a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ for both RWs</span>, with their statistical uncertainties, as function of the step $i\in [0,10^2]$.

Suggestion: divide your $M$ throws into a reasonable number of blocks, compute $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ as function of the step $i\in [0,10^2]$ in each block and use these values (for fixed $i$) to obtain the average value and its statistical uncertainty.

Note that you could try to fit both results with a function like $f(N)=k\sqrt{N}$. Do your results indicate a diffusive behavior?

### 3D Random Walks (RW) on a cubic lattice
-Number of RW $10^4$

-Number of blocks $10^2$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Read data from file
data = np.genfromtxt("Positions_RW1_try.out")

# Extract values and uncertainties
blocks = data[:, 0]  # Number of blocks
position_values = data[:, 1]  # Posistion values
uncertainties = data[:, 2]  # Uncertainties

# Plotting
print("Lattice constant a = 1")
plt.figure(figsize=(12,8))
plt.errorbar(blocks, position_values, yerr=uncertainties, fmt='o', markersize=3, capsize=5, color='g')
plt.xlabel('Step i')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.title('Cubic lattice RW')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definizione della funzione radice quadrata
def sqrt_function(N, k):
    return k * np.sqrt(N)

# Lettura dei dati dal file
data = np.genfromtxt("Positions_RW1_try.out")

# Estrazione dei valori e delle incertezze
blocks = data[:, 0]  # Numero di blocchi
position_values = data[:, 1]  # Valori delle posizioni
uncertainties = data[:, 2]  # Incertezze

# Esecuzione del fitting della curva
popt, pcov = curve_fit(sqrt_function, blocks, position_values)

# Estrazione del parametro ottimale
k = popt[0]

# Generazione dei valori fitted
fitted_values = sqrt_function(blocks, k)

# Stampa del parametro del fit
print(f"fit parameter: k = {k}")

# Plotting dei dati e del fit
plt.figure(figsize=(12,8))
plt.errorbar(blocks, position_values, yerr=uncertainties, fmt='o', markersize=5, capsize=3, color='g', label='Dati')
plt.plot(blocks, fitted_values, label=f'Fit: {k:.2e}*sqrt(N)', color='red', linewidth=5)
plt.xlabel('Step i')
plt.ylabel(r'$\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$')
plt.title('Cubic lattice RW fitted')
plt.legend()
plt.grid(True)
plt.show()
```
