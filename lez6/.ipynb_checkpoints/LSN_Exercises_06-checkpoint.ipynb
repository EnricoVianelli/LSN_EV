In Statistical Mechanics the one-dimensional (1D) Ising Model is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$:

- the **Internal Energy** per spin (with $h=0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e)
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.show()
```

- the **Heat Capacity** (with $h=0$):

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

```python
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
plt.plot(T, heat)
plt.title('Ising 1D, Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.show()
```

- The **magnetization** (with $h \not= 0$):

Note that with $h=0$ the magnetization $M(N,T,h=0)=\langle \sum_{i=1,N}s_i \rangle =0$ because in 1D we cannot have a (ferromagnetic) phase transition in absence of an external magnetic field.

$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$

The exact formula from $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ is:
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
plt.plot(T, M)
plt.title('Ising 1D, magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')
plt.show()
```

- The **magnetic susceptibility** (with $h=0$):

$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
plt.plot(T, X)
plt.title('Ising 1D, Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.show()
```

- $\rho = 1.0$
- $N = 50$
- $T = 0.5$
- $J = 1.0$ and $h = 0.0$
- number of Blocks: 20
- number of Steps: 20000

```python
import matplotlib.pyplot as plt
import numpy as np

def leggi_file(nome_file):
    blocchi = []
    medie = []
    errori = []
    
    with open(nome_file, 'r') as file:
        next(file)  # Ignora la prima riga
        for linea in file:
            blocchi_val, media_val, media_blocchi_val, errore_val = map(float, linea.strip().split())
            blocchi.append(blocchi_val*20000)
            medie.append(media_blocchi_val)  # Utilizziamo la media sui blocchi
            errori.append(errore_val)
    
    return blocchi, medie, errori

def confronta_metodologie(variabili, titoli_assi, etichette):
    num_variabili = len(variabili)
    plt.figure(figsize=(10, 15))  # Imposta la dimensione della figura
    
    for i in range(num_variabili):
        plt.subplot(num_variabili, 1, i+1)
        file1, file2 = variabili[i]
        blocchi1, medie1, errori1 = leggi_file(file1)
        blocchi2, medie2, errori2 = leggi_file(file2)

        plt.errorbar(blocchi1, medie1, yerr=errori1, label=etichette[i][1], fmt='o-', capsize=5, color='b')
        plt.errorbar(blocchi2, medie2, yerr=errori2, label=etichette[i][2], fmt='o-', capsize=5, color='r')
        plt.xlabel(titoli_assi[i][0])
        plt.ylabel(titoli_assi[i][1])
        plt.title(etichette[i][0])
        plt.legend()
        plt.grid(True)
        # Calcola i limiti dell'asse y basati sui dati con un certo margine
        tutte_medie = medie1 + medie2
        tutte_errori = errori1 + errori2
        min_y = min(tutte_medie) - max(tutte_errori)
        max_y = max(tutte_medie) + max(tutte_errori)
        plt.ylim(min_y - 0.1*(max_y-min_y), max_y + 0.1*(max_y-min_y))
        
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Per ciascuna variabile, fornisci i nomi dei file relativi alle due metodologie
    variabile1 = ('Gibbs_sampler/OUTPUT_Equilibration/OUTPUT/total_energy.dat', 
                  'Metropolis_sampler/OUTPUT_Equilibration/OUTPUT/total_energy.dat')
    variabile2 = ('Gibbs_sampler/OUTPUT_Equilibration/OUTPUT/magnetization.dat', 
                  'Metropolis_sampler/OUTPUT_Equilibration/OUTPUT/magnetization.dat')
    variabile3 = ('Gibbs_sampler/OUTPUT_Equilibration/OUTPUT/susceptibility.dat',
                  'Metropolis_sampler/OUTPUT_Equilibration/OUTPUT/susceptibility.dat')
    variabile4 = ('Gibbs_sampler/OUTPUT_Equilibration/OUTPUT/specific_heat.dat',
                  'Metropolis_sampler/OUTPUT_Equilibration/OUTPUT/specific_heat.dat')
    
    # Definisci i titoli degli assi per ciascuna variabile
    titoli_assi = [
        ('MC step', 'U/N'),
        ('MC step', 'M'),
        ('MC step', 'Chi'),
        ('MC step', 'CV')
    ]
    
    # Etichette per ciascuna variabile
    etichette = [
        ('Internal energy', 'Gibbs', 'Metropolis'),
        ('Magnetization', 'Gibbs', 'Metropolis'),
        ('Susceptibility', 'Gibbs', 'Metropolis'),
        ('Specific Heat', 'Gibbs', 'Metropolis')
    ]

    confronta_metodologie([variabile1, variabile2, variabile3, variabile4], titoli_assi, etichette)
```

## <span style="color:blue"> Simulation ising 1D</span> 

- $\rho = 1.0$
- $J = 1.0$ and $N = 50$
- number of Blocks: 20
- number of Steps: 20000

```python
import matplotlib.pyplot as plt
import numpy as np

# =========================
# Simulation Data Parameters
# =========================

# Filenames for simulation data
filenames = [
    [
        'Metropolis_sampler/T-0.25/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-0.5/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-0.75/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-1.0/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-1.25/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-1.5/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-1.75/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-2.0/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-2.25/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-2.5/OUTPUT/total_energy.dat'
    ],
    [
        'Metropolis_sampler/T-0.25/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-0.5/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-0.75/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-1.0/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-1.25/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-1.5/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-1.75/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-2.0/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-2.25/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-2.5/OUTPUT/specific_heat.dat'  # Corrected filename
    ],
    [
        'Metropolis_sampler/T-0.25_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-0.5_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-0.75_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-1.0_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-1.25_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-1.5_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-1.75_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-2.0_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-2.25_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-2.5_0.02/OUTPUT/magnetization.dat'
    ],
    [
        'Metropolis_sampler/T-0.25/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-0.5/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-0.75/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-1.0/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-1.25/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-1.5/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-1.75/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-2.0/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-2.25/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-2.5/OUTPUT/susceptibility.dat'
    ]
]

# Titles and labels for the subplots
titles = [
    'Internal Energy (Metropolis vs Analytical)',
    'Specific Heat (Metropolis vs Analytical)',
    'Magnetization M (h = 0.02) (Metropolis vs Analytical)',
    'Susceptibility (Metropolis vs Analytical)'
]

y_labels = [
    'U/N',
    'C',
    'M',
    'χ'
]

# Temperature range for simulation data
x_sim = np.linspace(0.25, 2.5, 10)

# =========================
# Analytical Solution Parameters
# =========================

# Analytical temperature range (for smooth curves)
points = 500
T_analytical = np.linspace(0.2, 3.0, num=points)
beta = 1.0 / T_analytical
J = 1.0  # Interaction constant
Ns = 50  # Number of spins
h = 0.02  # External magnetic field

# Precompute hyperbolic functions
th = np.tanh(J * beta)
thN = th ** Ns
ch = 1.0 / th

# Internal Energy U/N
e = -J * (th + ch * thN) / (1 + thN)

# Specific Heat C
heat = (beta * J) ** 2 * (((1 + thN + (Ns - 1) * (th ** 2) + (Ns - 1) * (ch ** 2) * thN) / (1 + thN)) -
                           Ns * ((th + ch * thN) / (1 + thN)) ** 2)

# Magnetization M with h = 0.02
# To avoid division by zero or numerical instability, add a small epsilon if needed
epsilon = 1e-10
sqrt_term = np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) ** 2 - 2 * np.sinh(2 * beta * J))
# Handle potential negative values under the square root due to numerical errors
sqrt_term = np.where(sqrt_term < 0, 0, sqrt_term)

l1 = np.exp(beta * J) * np.cosh(beta * h) + sqrt_term
l2 = np.exp(beta * J) * np.cosh(beta * h) - sqrt_term
Z = l1 ** Ns + l2 ** Ns

# Avoid division by zero in magnetization
Z = np.where(Z == 0, epsilon, Z)

# Magnetization formula (Corrected)
M = (np.exp(beta * J) * np.sinh(beta * h) *
     (l1 ** (Ns - 1) * (1 + np.exp(beta * J) * np.cosh(beta * h) / 
         np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) ** 2 - 2 * np.sinh(2 * beta * J))) +
      l2 ** (Ns - 1) * (1 - np.exp(beta * J) * np.cosh(beta * h) / 
         np.sqrt(np.exp(2 * beta * J) * np.cosh(beta * h) ** 2 - 2 * np.sinh(2 * beta * J))))) / (Z)

# Susceptibility χ
X = beta * np.exp(2 * beta * J) * (1 - thN) / (1 + thN)

# =========================
# Plotting
# =========================

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()  # Flatten to 1D array for easy iteration

# Iterate over each subplot
for i, (ax, title, y_label, file_group) in enumerate(zip(axes, titles, y_labels, filenames)):
    means = []
    errors = []
    
    # Read simulation data
    for filename in file_group:
        try:
            with open(filename, 'r') as file:
                lines = file.readlines()
                last_line = lines[-1].strip().split()
                # Adjust indices based on your data file format
                # Assuming the third column is the mean and the fourth is the error
                if len(last_line) >= 4:
                    means.append(float(last_line[2]))
                    errors.append(float(last_line[3]))
                else:
                    # If there are fewer columns, handle accordingly
                    means.append(float(last_line[-2]))
                    errors.append(float(last_line[-1]))
        except Exception as e:
            print(f"Error reading {filename}: {e}")
            means.append(np.nan)
            errors.append(np.nan)
    
    # Plot simulation data with error bars
    ax.errorbar(x_sim, means, yerr=errors, fmt='o', capsize=5, capthick=2,color='black', ecolor='red',
                label='Metropolis Simulation')
    
    # Plot analytical curve
    if i == 0:
        # Internal Energy
        ax.plot(T_analytical, e, linestyle='-', color='blue', label='Analytical')
    elif i == 1:
        # Specific Heat
        ax.plot(T_analytical, heat/50, linestyle='-', color='blue', label='Analytical')
    elif i == 2:
        # Magnetization
        ax.plot(T_analytical, M, linestyle='-', color='blue', label='Analytical')
    elif i == 3:
        # Susceptibility
        ax.plot(T_analytical, X, linestyle='-', color='blue', label='Analytical')
    
    # Customize subplot
    ax.set_title(title, fontsize=14)
    ax.set_xlabel('Temperature (T)', fontsize=12)
    ax.set_ylabel(y_label, fontsize=12)
    ax.legend(fontsize=12)
    ax.grid(True)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# =========================
# Parametri dei Dati di Simulazione
# =========================

# Lista delle temperature (assicurati che l'ordine corrisponda ai file)
temperatures = [0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0, 2.25, 2.5]
x_sim = np.array(temperatures)

# File di dati per Metropolis sampler
filenames_metropolis = [
    [  # Internal Energy
        'Metropolis_sampler/T-0.25/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-0.5/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-0.75/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-1.0/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-1.25/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-1.5/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-1.75/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-2.0/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-2.25/OUTPUT/total_energy.dat',
        'Metropolis_sampler/T-2.5/OUTPUT/total_energy.dat'
    ],
    [  # Specific Heat
        'Metropolis_sampler/T-0.25/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-0.5/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-0.75/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-1.0/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-1.25/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-1.5/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-1.75/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-2.0/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-2.25/OUTPUT/specific_heat.dat',
        'Metropolis_sampler/T-2.5/OUTPUT/specific_heat.dat'
    ],
    [  # Magnetization
        'Metropolis_sampler/T-0.25_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-0.5_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-0.75_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-1.0_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-1.25_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-1.5_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-1.75_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-2.0_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-2.25_0.02/OUTPUT/magnetization.dat',
        'Metropolis_sampler/T-2.5_0.02/OUTPUT/magnetization.dat'
    ],
    [  # Susceptibility
        'Metropolis_sampler/T-0.25/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-0.5/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-0.75/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-1.0/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-1.25/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-1.5/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-1.75/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-2.0/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-2.25/OUTPUT/susceptibility.dat',
        'Metropolis_sampler/T-2.5/OUTPUT/susceptibility.dat'
    ]
]

# File di dati per Gibbs sampler
filenames_gibbs = [
    [  # Internal Energy
        'Gibbs_sampler/T-0.25/OUTPUT/total_energy.dat',
        'Gibbs_sampler/T-0.5/OUTPUT/total_energy.dat',
        'Gibbs_sampler/T-0.75/OUTPUT/total_energy.dat',
        'Gibbs_sampler/T-1.0/OUTPUT/total_energy.dat',
        'Gibbs_sampler/T-1.25/OUTPUT/total_energy.dat',
        'Gibbs_sampler/T-1.5/OUTPUT/total_energy.dat',
        'Gibbs_sampler/T-1.75/OUTPUT/total_energy.dat',
        'Gibbs_sampler/T-2.0/OUTPUT/total_energy.dat',
        'Gibbs_sampler/T-2.25/OUTPUT/total_energy.dat',
        'Gibbs_sampler/T-2.5/OUTPUT/total_energy.dat'
    ],
    [  # Specific Heat
        'Gibbs_sampler/T-0.25/OUTPUT/specific_heat.dat',
        'Gibbs_sampler/T-0.5/OUTPUT/specific_heat.dat',
        'Gibbs_sampler/T-0.75/OUTPUT/specific_heat.dat',
        'Gibbs_sampler/T-1.0/OUTPUT/specific_heat.dat',
        'Gibbs_sampler/T-1.25/OUTPUT/specific_heat.dat',
        'Gibbs_sampler/T-1.5/OUTPUT/specific_heat.dat',
        'Gibbs_sampler/T-1.75/OUTPUT/specific_heat.dat',
        'Gibbs_sampler/T-2.0/OUTPUT/specific_heat.dat',
        'Gibbs_sampler/T-2.25/OUTPUT/specific_heat.dat',
        'Gibbs_sampler/T-2.5/OUTPUT/specific_heat.dat'
    ],
    [  # Magnetization
        'Gibbs_sampler/T-0.25_0.02/OUTPUT/magnetization.dat',
        'Gibbs_sampler/T-0.5_0.02/OUTPUT/magnetization.dat',
        'Gibbs_sampler/T-0.75_0.02/OUTPUT/magnetization.dat',
        'Gibbs_sampler/T-1.0_0.02/OUTPUT/magnetization.dat',
        'Gibbs_sampler/T-1.25_0.02/OUTPUT/magnetization.dat',
        'Gibbs_sampler/T-1.5_0.02/OUTPUT/magnetization.dat',
        'Gibbs_sampler/T-1.75_0.02/OUTPUT/magnetization.dat',
        'Gibbs_sampler/T-2.0_0.02/OUTPUT/magnetization.dat',
        'Gibbs_sampler/T-2.25_0.02/OUTPUT/magnetization.dat',
        'Gibbs_sampler/T-2.5_0.02/OUTPUT/magnetization.dat'
    ],
    [  # Susceptibility
        'Gibbs_sampler/T-0.25/OUTPUT/susceptibility.dat',
        'Gibbs_sampler/T-0.5/OUTPUT/susceptibility.dat',
        'Gibbs_sampler/T-0.75/OUTPUT/susceptibility.dat',
        'Gibbs_sampler/T-1.0/OUTPUT/susceptibility.dat',
        'Gibbs_sampler/T-1.25/OUTPUT/susceptibility.dat',
        'Gibbs_sampler/T-1.5/OUTPUT/susceptibility.dat',
        'Gibbs_sampler/T-1.75/OUTPUT/susceptibility.dat',
        'Gibbs_sampler/T-2.0/OUTPUT/susceptibility.dat',
        'Gibbs_sampler/T-2.25/OUTPUT/susceptibility.dat',
        'Gibbs_sampler/T-2.5/OUTPUT/susceptibility.dat'
    ]
]

# Titoli e etichette per i grafici (ordine corrispondente alle liste sopra)
titles = [
    'Internal Energy (Metropolis vs Gibbs)',
    'Specific Heat (Metropolis vs Gibbs)',
    'Magnetization M (h = 0.02) (Metropolis vs Gibbs)',
    'Susceptibility (Metropolis vs Gibbs)'
]

y_labels = [
    'U/N',
    'C',
    'M',
    'χ'
]

# =========================
# Funzione per Leggere i Dati di Simulazione
# =========================

def read_simulation_data(file_list):
    """
    Legge i dati di simulazione da una lista di file.

    Parameters:
    - file_list: lista di stringhe, percorsi ai file di dati.

    Returns:
    - means: lista di float, medie delle quantità.
    - errors: lista di float, errori delle quantità.
    """
    means = []
    errors = []
    for filename in file_list:
        try:
            with open(filename, 'r') as file:
                lines = file.readlines()
                last_line = lines[-1].strip().split()
                # Supponendo che la terza colonna sia la media e la quarta l'errore
                if len(last_line) >= 4:
                    means.append(float(last_line[2]))
                    errors.append(float(last_line[3]))
                else:
                    # Se ci sono meno colonne, prendere gli ultimi due valori
                    means.append(float(last_line[-2]))
                    errors.append(float(last_line[-1]))
        except Exception as e:
            print(f"Errore nella lettura del file {filename}: {e}")
            means.append(np.nan)
            errors.append(np.nan)
    return means, errors

# =========================
# Lettura dei Dati di Simulazione
# =========================

# Inizializzazione dizionari per Metropolis e Gibbs
simulation_data_metropolis = {}
simulation_data_gibbs = {}

quantities = ['Internal Energy', 'Specific Heat', 'Magnetization', 'Susceptibility']

# Lettura dei dati per Metropolis
for i, quantity in enumerate(quantities):
    means, errors = read_simulation_data(filenames_metropolis[i])
    simulation_data_metropolis[quantity] = {'means': means, 'errors': errors}

# Lettura dei dati per Gibbs
for i, quantity in enumerate(quantities):
    means, errors = read_simulation_data(filenames_gibbs[i])
    simulation_data_gibbs[quantity] = {'means': means, 'errors': errors}

# =========================
# Plotting
# =========================

# Creazione dei grafici
fig, axes = plt.subplots(2, 2, figsize=(20, 15))
axes = axes.flatten()  # Per iterare facilmente sugli assi

# Iterazione su ciascuna quantità
for i, (quantity, ax) in enumerate(zip(quantities, axes)):
    # Dati Metropolis
    met_means = simulation_data_metropolis[quantity]['means']
    met_errors = simulation_data_metropolis[quantity]['errors']
    
    # Dati Gibbs
    gib_means = simulation_data_gibbs[quantity]['means']
    gib_errors = simulation_data_gibbs[quantity]['errors']
    
    # Plot Metropolis
    ax.errorbar(
        x_sim, met_means, yerr=met_errors, fmt='o', capsize=5, capthick=2,
        ecolor='red', color='red', label='Metropolis Simulation',
        linestyle='-', markersize=8  # Rimuove 'marker' keyword
    )
    
    # Plot Gibbs
    ax.errorbar(
        x_sim, gib_means, yerr=gib_errors, fmt='s', capsize=5, capthick=2,
        ecolor='green', color='green', label='Gibbs Simulation',
        linestyle='--', markersize=8  # Rimuove 'marker' keyword
    )
    
    # Personalizzazione del grafico
    ax.set_title(titles[i], fontsize=16)
    ax.set_xlabel('Temperature (T)', fontsize=14)
    ax.set_ylabel(y_labels[i], fontsize=14)
    ax.legend(fontsize=12)
    ax.grid(True)

# Aggiustamenti finali
plt.tight_layout()
plt.show()
```
