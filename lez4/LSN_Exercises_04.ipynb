\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] \right\rangle $$

### <span style="color:green"> Equilibration is needed!</span>

By monitoring the temperature, $T$, during the first time steps of the simulation, you will observe that at the beginning, starting from a perfect crystalline configuration, $T$ will not conserve its value. Can you understand and explain this behaviour?

Explore and find the starting value of $T$ needed to stabilize, after a certain number of time steps (equilibration time), the average value of $T$ on the desired $T^\star$. <span style="color:red">Perform your solid/liquid/gas phase MD simulations by re-starting them after such equilibration time.</span>

<span style="color:red">Solid equilibration:</span> 25 blocks of 2000 steps and $T = 1.541$.

<span style="color:blue">Liquid equilibration:</span> 25 blocks of 2000 steps and $T = 1.985$.

<span style="color:green">Gas equilibration:</span> 30 blocks of 2000 steps and $T = 0.972$.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to read and process data
def read_and_process_data(file_path):
    data = pd.read_csv(file_path, delim_whitespace=True)
    if '# BLOCK:' in data.columns[0]:
        data = data.iloc[1:].reset_index(drop=True)
    data.columns = ['BLOCK', 'ACTUAL_T', 'T_AVE', 'ERROR', 'EXTRA'] if data.shape[1] == 5 else ['BLOCK', 'ACTUAL_T', 'T_AVE', 'ERROR']
    return data[['BLOCK', 'ACTUAL_T', 'T_AVE', 'ERROR']].apply(pd.to_numeric)

# File paths for Solid, Liquid, and Gas
file_paths = {
    'Solid': 'Solid/OUTPUT_equilibration/OUTPUT/temperature.dat',
    'Liquid': 'Liquid/OUTPUT_equilibration/OUTPUT/temperature.dat',
    'Gas': 'Gas/OUTPUT_equilibration/OUTPUT/temperature.dat'
}

# Initialize subplots
fig, axes = plt.subplots(3, 1, figsize=(10, 18), sharex=True)
colors = ['r', 'b', 'g']

# Process and plot data for each phase
for ax, (phase, file_path), color in zip(axes, file_paths.items(), colors):
    data = read_and_process_data(file_path)
    num_blocchi = data['BLOCK']
    medie_blocchi = data['T_AVE']
    errori = data['ERROR']

    ax.errorbar(num_blocchi, medie_blocchi, yerr=errori, fmt='-o', capsize=5, color=color, label=f'{phase} temperature.dat')
    ax.set_xlabel('# of blocks')
    ax.set_ylabel('T')
    ax.set_title(f'Equilibration of {phase}')
    ax.legend()
    ax.grid(True)
    ax.set_xticks(num_blocchi[::2])
    ax.set_xticklabels(num_blocchi[::2])

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

## <span style="color:red"> Solid phase</span> 

- $\rho^\star = 1.1$
- $T^\star = 0.8$
- cut-off radius: $r_c = 2.2$
- number of blocks: 100
- number of steps: 2000

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator

def leggi_file(nome_file):
    # Inizializza le liste per memorizzare i dati
    num_blocchi = []
    media_blocchi = []
    media_totale = []
    errore = []

    # Apre il file e legge i dati
    with open(nome_file, 'r') as file:
        # Ignora l'intestazione
        next(file)
        for riga in file:
            valori = riga.strip().split()
            num_blocchi.append(int(valori[0]))
            media_blocchi.append(float(valori[1]))
            media_totale.append(float(valori[2]))
            errore.append(float(valori[3]))

    return num_blocchi, media_blocchi, media_totale, errore
    
def crea_grafico(nome_file, num_blocchi, media_totale, errore, colore, titolo, titolo_y):
    plt.figure(figsize=(12,7))
    plt.errorbar(num_blocchi, media_totale, yerr=errore, fmt='o-', label=nome_file, color=colore, capsize = 2)
    plt.xlabel('Number of blocks')
    plt.ylabel(titolo_y)
    plt.title(titolo)
    plt.xticks(num_blocchi)
    plt.gca().xaxis.set_major_locator(MultipleLocator(5))  # Imposta le etichette ogni 5 punti sull'asse x
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    # Elenco dei nomi dei file, dei colori e dei titoli
    file_info = [
        ('Solid/OUTPUT/total_energy.dat', 'blue', 'Total energy per particle', 'E/N'),
        ('Solid/OUTPUT/kinetic_energy.dat', 'red', 'Kinetic energy per particle', 'K/N'),
        ('Solid/OUTPUT/potential_energy.dat', 'green', 'Potential energy per particle', 'U/N'),
        ('Solid/OUTPUT/temperature.dat', 'orange', 'Temperature', 'T'),
        ('Solid/OUTPUT/pressure.dat', 'purple', 'Pressure', 'P')]
    for file_name, colore, titolo, titolo_y in file_info:
       # Legge i dati dal file
        num_blocchi, _, media_totale, errore = leggi_file(file_name)

        # Crea il grafico
        crea_grafico(file_name, num_blocchi, media_totale, errore, colore, titolo, titolo_y)

if __name__ == "__main__":
    main()
```
